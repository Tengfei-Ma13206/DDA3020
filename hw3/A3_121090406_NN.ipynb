```python
from sklearn.neural_network import MLPClassifier
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv',nrows=6000).dropna()
test_data = pd.read_csv('test.csv',nrows=1000).dropna()
X_train = train_data.iloc[:,2:].values / 255
y_train = train_data['Category'].values
X_test = test_data.iloc[:,1:].values / 255
y_test = test_data['label'].values
```

```python
train_data
```

```python
result = []
for numHiddenLayers in (1,2,3):
    for numHiddenNodes in (50,200,784):
        for optimizer in ('adam','sgd'):
            for learningRate in (0.001,0.01,0.1):
                clf = MLPClassifier(hidden_layer_sizes=(numHiddenNodes,)*numHiddenLayers,solver=optimizer,learning_rate_init=learningRate,early_stopping=True).fit(X_train,y_train)
                train_accuracy = clf.score(X_train,y_train)
                test_accuracy = clf.score(X_test,y_test)
                result.append({'numHiddenLayers':numHiddenLayers,'numHiddenNodes':numHiddenNodes,'optimizer':optimizer,'learningRate':learningRate,'train_accuracy':train_accuracy,'test_accuracy':test_accuracy})
```
